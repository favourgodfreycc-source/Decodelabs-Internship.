```python
import pandas as pd
import os

# Get the current working directory of the notebook
print(f"Current directory: {os.getcwd()}")

# Load the file using the full path
# Since your terminal shows you are in 'Data_Analytics_Project_1'
file_path = 'Data_Analytics_Project_1/Dataset for Data Analytics.xlsx'
df = pd.read_excel(file_path)

df.head()
```

```python
# 1. Check for missing values (The gaps in the brick wall)
print("Missing Values:\n", df.isnull().sum())

# 2. Check for duplicates (The redundant records)
print(f"\nNumber of duplicate OrderIDs: {df['OrderID'].duplicated().sum()}")
```

```python
# Fill the missing CouponCode values with the most common code (the Mode)
most_common_coupon = df['CouponCode'].mode()[0]
df['CouponCode'] = df['CouponCode'].fillna(most_common_coupon)

# Verify the fix
print("Missing values after imputation:\n", df['CouponCode'].isnull().sum())
```

```python
# 1. Standardize Dates to YYYY-MM-DD
df['Date'] = pd.to_datetime(df['Date']).dt.strftime('%Y-%m-%d')

# 2. Standardize Numeric Precision to 2 decimal places
df['UnitPrice'] = df['UnitPrice'].round(2)
df['TotalPrice'] = df['TotalPrice'].round(2)
```

```python
# Check for duplicates
print(f"Duplicates before removal: {df.duplicated().sum()}")

# Remove duplicates
df = df.drop_duplicates()

# Verify removal
print(f"Duplicates after removal: {df.duplicated().sum()}")
```

```python
df['Date'] = pd.to_datetime(df['Date']).dt.normalize() 
```

```python
# Check for nulls in the whole dataset
print("Missing values in dataset:\n", df.isnull().sum())

# Final check for duplicates
print("Are there any duplicates left?", df.duplicated().any())
```
